# Waste generated Europe

What is the total waste generated in Europe across the years?

```python
query='''
SELECT  
    SUM(`2011`) AS Total_2011,
    SUM(`2012`) AS Total_2012,
    SUM(`2013`) AS Total_2013,
    SUM(`2014`) AS Total_2014,
    SUM(`2015`) AS Total_2015,
    SUM(`2016`) AS Total_2016,
    SUM(`2017`) AS Total_2017,
    SUM(`2018`) AS Total_2018,
    SUM(`2019`) AS Total_2019,
    SUM(`2020`) AS Total_2020
FROM df_generated_global_packaging
WHERE COUNTRY <> 'European Union - 27 countries (from 2020)'
'''
pd.read_sql_query (query, engine)
```

What are the 10 countries with more garbagge generated per capita?

```python
query='''
SELECT COUNTRY, SUM(`2011` + `2012` + `2013` + `2014` + `2015` + `2016` + `2017` + `2018` + `2019` + `2020`) AS Total_Generation
FROM df_generated_global_packaging
GROUP BY COUNTRY
ORDER BY Total_Generation DESC
LIMIT 10;
'''
pd.read_sql_query (query, engine)
```

# Recycling ratio

Which is the average total ratio of recyling per year in europe?

```python
query = '''
SELECT
    AVG(`2011`) AS Avg_ratio_2011,
    AVG(`2012`) AS Avg_ratio_2012,
    AVG(`2013`) AS Avg_ratio_2013,
    AVG(`2014`) AS Avg_ratio_2014,
    AVG(`2015`) AS Avg_ratio_2015,
    AVG(`2016`) AS Avg_ratio_2016,
    AVG(`2017`) AS Avg_ratio_2017,
    AVG(`2018`) AS Avg_ratio_2018,
    AVG(`2019`) AS Avg_ratio_2019,
    AVG(`2020`) AS Avg_ratio_2020
FROM
    recy_ratio_europe;
'''

pd.read_sql_query(query, engine)
```

```python
query = '''
SELECT
    COUNTRY,
    AVG(`2011`) AS Avg_ratio_2011,
    AVG(`2012`) AS Avg_ratio_2012,
    AVG(`2013`) AS Avg_ratio_2013,
    AVG(`2014`) AS Avg_ratio_2014,
    AVG(`2015`) AS Avg_ratio_2015,
    AVG(`2016`) AS Avg_ratio_2016,
    AVG(`2017`) AS Avg_ratio_2017,
    AVG(`2018`) AS Avg_ratio_2018,
    AVG(`2019`) AS Avg_ratio_2019,
    AVG(`2020`) AS Avg_ratio_2020
FROM
    recy_ratio_europe
GROUP BY
    COUNTRY
ORDER BY
    Avg_ratio_2011 DESC
LIMIT 10;
'''

pd.read_sql_query(query, engine)
```

What are the top 9 countries with higher ratio over the years?

```python
query='''
SELECT COUNTRY, MAX(`2011`) AS Max_2011, MAX(`2012`) AS Max_2012, MAX(`2013`) AS Max_2013, MAX(`2014`) AS Max_2014, MAX(`2015`) AS Max_2015, MAX(`2016`) AS Max_2016, MAX(`2017`) AS Max_2017, MAX(`2018`) AS Max_2018, MAX(`2019`) AS Max_2019, MAX(`2020`) AS Max_2020
FROM recy_ratio_europe
GROUP BY COUNTRY
ORDER BY MAX(`2011` + `2012` + `2013` + `2014` + `2015` + `2016` + `2017` + `2018` + `2019` + `2020`) DESC
LIMIT 10;
'''
pd.read_sql_query (query, engine)
```

3 Countries with a ratio improvement across the years

```python
query='''
SELECT
    Country,
    (`2012` - `2011`) AS Increment_2012,
    (`2013` - `2012`) AS Increment_2013,
    (`2014` - `2013`) AS Increment_2014,
    (`2015` - `2014`) AS Increment_2015,
    (`2016` - `2015`) AS Increment_2016,
    (`2017` - `2016`) AS Increment_2017,
    (`2018` - `2017`) AS Increment_2018,
    (`2019` - `2018`) AS Increment_2019,
    (`2020` - `2019`) AS Increment_2020
FROM recy_ratio_europe
ORDER BY (Increment_2012 + Increment_2013 + Increment_2014 + Increment_2015 + Increment_2016 + Increment_2017 + Increment_2018 + Increment_2019 + Increment_2020) DESC
LIMIT 3
'''
pd.read_sql_query (query, engine)
```

```python
query='''
SELECT
    Country,
    ((`2012` - `2011`) / `2011` * 100) AS Increment_2012,
    ((`2013` - `2012`) / `2012` * 100) AS Increment_2013,
    ((`2014` - `2013`) / `2013` * 100) AS Increment_2014,
    ((`2015` - `2014`) / `2014` * 100) AS Increment_2015,
    ((`2016` - `2015`) / `2015` * 100) AS Increment_2016,
    ((`2017` - `2016`) / `2016` * 100) AS Increment_2017,
    ((`2018` - `2017`) / `2017` * 100) AS Increment_2018,
    ((`2019` - `2018`) / `2018` * 100) AS Increment_2019,
    ((`2020` - `2019`) / `2019` * 100) AS Increment_2020
FROM recy_ratio_europe
ORDER BY (Increment_2012 + Increment_2013 + Increment_2014 + Increment_2015 + Increment_2016 + Increment_2017 + Increment_2018 + Increment_2019 + Increment_2020) DESC
LIMIT 3
'''
pd.read_sql_query (query, engine)
```
